```python
#basic ledger and chain class
class Ledger():
    
    def __init__(self,accounts,prev_hash):
        self.accounts=accounts
        self.prev_hash=prev_hash
        self.hash=None
        self.prev_block=None
    

class Chain():
    def __init__(self,initiation,blocks,pos):
        self.initiation=initiation
        self.blocks=blocks
        self.pos=pos
        
        
    def Add_Block(self,acc,hss):
        self.blocks+=1
        if self.initiation==None:
            dummy=Ledger(acc,"77777")
            self.initiation=dummy
            self.pos=dummy
            dummy.hash=hss
            self.pos=dummy
        else:
            dummy=Ledger(acc,hss)
            dummy1=self.pos
            self.pos=dummy
            self.pos.hash=hss
            self.pos.prev_block=dummy1
            self.pos.prev_hash=dummy1.hash
            
myc=Chain(None,0,0)
```
